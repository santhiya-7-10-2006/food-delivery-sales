```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Order_ID': [1,2,3,4,5,6,7],
    'Order_Time': ['12:30', '13:00', '19:15', '20:00', '12:45', '19:30', '20:15'],
    'Location': ['Chennai', 'Madurai', 'Chennai', 'Coimbatore', 'Chennai', 'Madurai', 'Chennai'],
    'Item': ['Biryani', 'Dosa', 'Pizza', 'Burger', 'Biryani', 'Pizza', 'Burger'],
    'Price': [200,80,250,150,200,250,150]
}

df = pd.DataFrame(data)

print(df.head())
```

```python
food_count = df['Item'].value_counts()

plt.figure(figsize=(6,4))
food_count.plot(kind='bar')
plt.title("Popular Food Items")
plt.xlabel("Food Item")
plt.ylabel("Orders")
plt.show()
```

```python
df['Hour'] = pd.to_datetime(df['Order_Time'], format='%H:%M').dt.hour

hour_count = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(6,4))
hour_count.plot(kind='bar')
plt.title("Peak Order Time")
plt.xlabel("Hour")
plt.ylabel("Number of Orders")
plt.show()
```

```python
location_count = df['Location'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(location_count,
        labels=location_count.index,
        autopct='%1.1f%%')

plt.title("Location-wise Orders")
plt.show()
```

```python
most_popular = df['Item'].value_counts().idxmax()
peak_hour = df['Hour'].value_counts().idxmax()
top_location = df['Location'].value_counts().idxmax()
total_revenue = df['Price'].sum()

print("BUSINESS INSIGHTS")
print("Most Popular Food Item :", most_popular)
print("Peak Order Hour :", peak_hour)
print("Highest Order Location :", top_location)
print("Total Revenue : Rs.", total_revenue)
```
